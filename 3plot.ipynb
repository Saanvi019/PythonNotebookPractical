```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("customers-100.csv")

print(data.head())
```

```python
data["Subscription Date"] = pd.to_datetime(data["Subscription Date"])

data["Year"] = data["Subscription Date"].dt.year
data["Month"] = data["Subscription Date"].dt.month
```

```python
plt.figure()
data["Year"].hist()
plt.title("Subscription Year Distribution")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
top_countries = data["Country"].value_counts().head(5)

plt.figure()
top_countries.plot(kind="bar")
plt.title("Top 5 Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
top_countries.plot(kind="pie", autopct="%1.1f%%")
plt.title("Country Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure()
plt.scatter(data["Year"], data["Month"])
plt.title("Year vs Month")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
